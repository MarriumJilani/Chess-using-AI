```python
def get_positions(move):

    # coz using A5A6 kinda format
    r_from = 8 - int(move[1]) #row jahan se piece moved
    col_from = ord(move[0]) - ord('A') #col jahan se piece moved
    r_to = 8 - int(move[3]) #row to which piece moved
    col_to = ord(move[2]) - ord('A') #col to which piece moved
    return r_from, col_from, r_to, col_to

r_from, col_from, r_to, col_to=get_positions("E2E4")
print(r_from, col_from,r_to, col_to)
```

```python
def is_in_check(chessboard, from_row, from_col, to_row, to_col, color):

  # temporary copy to store 
  original_piece = chessboard[to_row][to_col]
  
  # move the piece to the destination square for checking
  chessboard[to_row][to_col] = chessboard[from_row][from_col]
  chessboard[from_row][from_col] = "   "
  
  king_pos = None
  if color == "W":
    # position of white King
    for i in range(len(chessboard)):
      for j in range(len(chessboard[i])):
        if chessboard[i][j] == "WKK":
          king_pos = (i, j)
  else:
    # position of  black King
    for i in range(len(chessboard)):
      for j in range(len(chessboard[i])):
        if chessboard[i][j] == "BKK":
          king_pos = (i, j)
  
  # check if the King is in check
  for i in range(len(chessboard)):
    for j in range(len(chessboard[i])):
      piece = chessboard[i][j]
      if piece != "   " and piece[0] != color[0]:
        #print(piece, i, j, king_pos[0], king_pos[1])
        if can_move(chessboard,piece, i, j, king_pos[0], king_pos[1]):
          # undo the move and return True if King is in check
          #print(piece, i, j, king_pos[0], king_pos[1])
          chessboard[from_row][from_col] = chessboard[to_row][to_col]
          chessboard[to_row][to_col] = original_piece
          return True
  
  # undo the move and return False if King is not in check
  chessboard[from_row][from_col] = chessboard[to_row][to_col]
  chessboard[to_row][to_col] = original_piece
  return False
 
################################################################################

def move_piece(chessboard, piece, fromto):
    r_from, col_from, r_to, col_to=get_positions(fromto)

    if chessboard[r_to][col_to][1]=="K" and chessboard[r_to][col_to][2]=="K":
      #print("CANT KILL KING")
      return False

    # check if the piece can move to the target position
    if can_move(chessboard, piece, r_from, col_from, r_to, col_to):
        # move the piece to the new position
        chessboard[r_to][col_to] = piece
        chessboard[r_from][col_from] = "   "
        return True
    else:
        return False

def can_move(chessboard, piece, from_row, from_col, to_row, to_col):
    # check if the target position is on the board
    if to_row < 0 or to_row > 7 or to_col < 0 or to_col > 7:
      return False

    # check if the piece is moving to its own position
    if from_row == to_row and from_col == to_col:
      return False

    # check if the piece is moving to a square occupied by its own color
    if chessboard[to_row][to_col][0] == piece[0]:
      #print("Its own color piece")
      return False
################################################################################    
    # check if the piece can legally move to the target position
    # PAWN CHECKKK
    if piece[1] == "P":
      
      if from_col == to_col:
        if chessboard[to_row][to_col] == "   ":
          if to_row == from_row + 1 and piece[0] == "B":
            return True
          elif to_row == from_row - 1 and piece[0] == "W":
            #print("here")
            return True
          elif from_row == 1 and to_row == 3 and chessboard[2][from_col] == "   ": #BLACK
            return True
          elif from_row == 6 and to_row == 4 and chessboard[5][from_col] == "   ": ##WHITE 
            return True

        else:
          return False

      elif abs(from_col - to_col) == 1 and to_row == from_row + 1:
        if chessboard[to_row][to_col] != "   ":
          #print("Can capture piece!")#can capture diagonally if wahan opponent ka piece ho
          return True
        else:
          return False
      else:
        #print("Invalid move for pawn!")   
        return False
###############################################################################
    # ROOK CHECKKKK
    elif piece[1] == "R": 

      if from_row == to_row:
        if from_col < to_col:
          for i in range(from_col + 1, to_col):
            if chessboard[from_row][i] != "   ":
              #print(chessboard[from_row][i]," is in the way")
              return False
          if chessboard[to_row][to_col][0] == piece[0]:
              return False #agar destination has same color piece tou invalid and block
          return True

        else:
          for i in range(to_col + 1, from_col):
            if chessboard[from_row][i] != "   ":
              #print(chessboard[from_row][i]," is in the way")
              #print(chessboard[from_row][i])
              return False
          if chessboard[to_row][to_col][0] == piece[0]:
            return False
          return True

      elif from_col == to_col:
        if from_row < to_row:
          for i in range(from_row + 1, to_row):
            if chessboard[i][from_col] != "   ":
              #print(chessboard[from_row][i]," is in the way")
              return False
          if chessboard[to_row][to_col][0] == piece[0]:
            return False
          return True
        else:
          for i in range(to_row + 1, from_row):
            if chessboard[i][from_col] != "   ":
              #print(chessboard[from_row][i]," is in the way")
              return False
          if chessboard[to_row][to_col][0] == piece[0]:
            return False
          return True

      else:
        return False
################################################################################
 # KNIGHT CHECKKK
    elif piece[1] == "K" and piece[2] != "K":

        if abs(from_row - to_row) == 2 and abs(from_col - to_col) == 1:
            return True
        elif abs(from_row - to_row) == 1 and abs(from_col - to_col) == 2:
            return True
        else:
            return False

################################################################################
  # BISHOP CHECKKK
    elif piece[1] == "B":
      if abs(from_row - to_row) != abs(from_col - to_col):
        return False

      else:
        if from_row < to_row and from_col < to_col:
          for i in range(1, to_row - from_row):
            if chessboard[from_row + i][from_col + i] != "   ":
              return False
          return True
        elif from_row < to_row and from_col > to_col:
          for i in range(1, to_row - from_row):
            if chessboard[from_row + i][from_col - i] != "   ":
              return False
          return True
        elif from_row > to_row and from_col < to_col:
          for i in range(1, from_row - to_row):
            if chessboard[from_row - i][from_col + i] != "   ":
              return False
          return True
        elif from_row > to_row and from_col > to_col:
          for i in range(1, from_row - to_row):
            if chessboard[from_row - i][from_col - i] != "   ":
              return False
          return True
        else:
            return False

################################################################################
# QUEEN CHECKKK
    elif piece[1] == "Q":
      #cross and square
      if from_row == to_row:
        if from_col < to_col:
          for i in range(from_col + 1, to_col):
            if chessboard[from_row][i] != "   ":
              return False
          return True
        else:
          for i in range(to_col + 1, from_col):
            if chessboard[from_row][i] != "   ":
              return False
          return True

      elif from_col == to_col:
        if from_row < to_row:
          for i in range(from_row + 1, to_row):
            if chessboard[i][from_col] != "   ":
              return False
          return True
        else:
          for i in range(to_row + 1, from_row):
            if chessboard[i][from_col] != "   ":
              return False
          return True

      elif abs(from_row - to_row) == abs(from_col - to_col):
        if from_row < to_row and from_col < to_col:
          for i in range(1, to_row - from_row):
            if chessboard[from_row + i][from_col + i] != "   ":
              return False
          return True
        elif from_row < to_row and from_col > to_col:
          for i in range(1, to_row - from_row):
            if chessboard[from_row + i][from_col - i] != "   ":
              return False
          return True
        elif from_row > to_row and from_col < to_col:
          for i in range(1, from_row - to_row):
            if chessboard[from_row - i][from_col + i] != "   ":
              return False
          return True
        elif from_row > to_row and from_col > to_col:
          for i in range(1, from_row - to_row):
            if chessboard[from_row - i][from_col - i] != "   ":
              return False
          return True
        else:
          return False
      else:
          return False
################################################################################
#KING CHECKKK
    elif piece[1] == "K" and  piece[2] == "K":
      #print("here")
      if abs(from_row - to_row) <= 1 and abs(from_col - to_col) <= 1:
        # check if the move puts the king in check
        if is_in_check(chessboard, from_row, from_col, to_row, to_col, piece[0]):
          #print("King in check")
          return False
        else:
          return True
      else:
        return False

################################################################################

chessboard1 = [   ["BR1", "   ", "BB1", "BQ ", "BKK", "BB2", "BK2", "BR2"],
                  ["   ", "   ", "   ", "BP4", "   ", "BP6", "   ", "BP8"],
                  ["   ", "BP2", "BK1", "   ", "   ", "   ", "   ", "   "],
                  ["BP1", "WB2", "   ", "   ", "WP5", "   ", "BP7", "   "],
                  ["   ", "   ", "WP4", "   ", "   ", "   ", "   ", "   "],
                  ["   ", "   ", "   ", "   ", "BP5", "   ", "   ", "   "],
                  ["WP1", "WP2", "WP3", "WKK", "BP3", "WP6", "WP7", "WP8"],
                  ["WR1", "WK1", "WB1", "WQ ", "   ", "   ", "WK2", "WR2"]
]

print_board(chessboard1)
r_from, col_from, r_to, col_to=get_positions("E3D2")
print(can_move(chessboard1,"BP5",r_from, col_from, r_to, col_to)) 
#print(move_piece(chessboard1, "BB1", "D5G1"))
move_piece(chessboard1, "BP5", "E3D2")
print_board(chessboard1)
```

```python
import copy
import math

def minimax_alpha_beta(chessboard, depth, alpha, beta, player):
    # base case
    if depth == 0:
        return None, evaluate_board(chessboard)

    # getting all possible moves for specific color player
    moves = get_all_moves(chessboard, player)

    best_move = None
    best_score = -float("inf") if player else float("inf")

    # for each move jo all possible moves se we get
    for move in moves:
        # making a copy 
        temp_chessboard = copy.deepcopy(chessboard)

        #print(move)
        move_piece(temp_chessboard, move[:3], move[3:])

        # recursively call minimax_alpha_beta
        _, score = minimax_alpha_beta(temp_chessboard, depth - 1, alpha, beta, not player)

        # update alpha or beta values accordingly
        if player:
            if score > best_score:
                best_score = score
                best_move = move
            alpha = max(alpha, best_score)
        else:
            if score < best_score:
                best_score = score
                best_move = move
            beta = min(beta, best_score)

        # performing alpha-beta pruning
        if alpha >= beta:
            break
    #print(best_move, best_score)
    return best_move, best_score



def get_all_moves(chessboard, player):
    moves = []

    # for each box
    for row in range(8):
        for col in range(8):
            # if the piece belongs to the current player
            if chessboard[row][col] != "   " and chessboard[row][col][0] == ("W" if player else "B"):
                # get all possible moves for the piece
                for move in get_possible_moves(chessboard, row, col):
                    piece=chessboard[row][col]
                    #print(piece)
                    moves.append(piece + move)

    return moves


def get_possible_moves(chessboard, row, col):
    piece = chessboard[row][col]
    moves = []
   
    # for each possible move
    for r in range(8):
        for c in range(8):
            if can_move(chessboard, piece, row, col, r, c):
                #print(piece)
                # Add move in the format of "from_position to_position"
                from_pos = chr(col + 65) + str(8 - row)
                to_pos = chr(c + 65) + str(8 - r)
                moves.append(from_pos + to_pos)

    return moves

################################################################################
def evaluate_board(chessboard):
    score = 0
    
    # see value of pieces
    white_material = 0
    black_material = 0
    for row in range(8):
      for col in range(8):
        piece = chessboard[row][col]
        if piece != "   ":
          if piece[1] == "P":
            value = 2
          elif piece[1] == "K" and piece[2] != "K": #knight
            value=3
          elif piece[1] == "B":
            value = 4
          elif piece[1] == "R":
            value = 5
          elif piece[1] == "Q":
            value = 6
          elif piece[1] == "K":
            # kings are worth infinite points kyunk they determine winning or losing
            value = 10

          if piece[0] == "W":  
            white_material += value
            #print("ehite",white_material)
          else:
            black_material += value
            #print("bl",black_material)
    
    # computer is white
    score += (white_material - black_material)
    
    if is_checkmate(chessboard,"W"):
        score += 100 

    # piece mobility and threat avoidance
    white_mobility = 0
    black_mobility = 0
    for row in range(8):
        for col in range(8):
            piece = chessboard[row][col]
            if piece != "   ":
                value = len(get_possible_moves(chessboard, row, col))
                if piece[0] == "W":
                    # check for possible moves that avoid opponent's threats
                    threat_count = count_threats(chessboard, row, col, "B")
                    if threat_count > 0:
                        avoidance_moves = get_avoidance_moves(chessboard, row, col, "B")
                        value += len(avoidance_moves)
                    white_mobility += value
                else:
                    # check for possible moves that avoid opponent's threats
                    threat_count = count_threats(chessboard, row, col, "W")
                    if threat_count > 0:
                        avoidance_moves = get_avoidance_moves(chessboard, row, col, "W")
                        value += len(avoidance_moves)
                    black_mobility += value

    #print(score)
    score += (white_mobility - black_mobility)

    # control of the center
    white_center_control = 0
    black_center_control = 0
    for row in range(2, 6):
        for col in range(2, 6):
            if chessboard[row][col] != "   ":
                piece = chessboard[row][col]
                if piece[0]=="W":
                    white_center_control += 1
                else:
                    black_center_control += 1

    score += (white_center_control - black_center_control)
                    
    #print("here",score)
    #print(score)
    return score



################################################################################

chessboard3 = [   ["BR1", "BK1", "BB1", "BQ ", "BKK", "BB2", "BK2", "BR2"],
                  ["BP1", "BP2", "   ", "BP4", "BP5", "BP6", "BP7", "BP8"],
                  ["   ", "   ", "   ", "   ", "   ", "   ", "   ", "   "],
                  ["   ", "   ", "   ", "   ", "   ", "   ", "   ", "   "],
                  ["   ", "   ", "   ", "BP3", "   ", "   ", "   ", "   "],
                  ["   ", "   ", "   ", "   ", "   ", "   ", "   ", "   "],
                  ["WP1", "WP2", "WP3", "WP4", "WP5", "WP6", "WP7", "WP8"],
                  ["WR1", "WK1", "WB1", "WQ ", "WKK", "WB2", "WK2", "WR2"]
]

def get_user_move():
    piece= input("Enter the piece you want to move (e.g. 'BP1'): ")
    move = input("Enter your move (e.g. 'A7A5'): ")
    return piece,move

def get_computer_move(board, depth):
    move, value = minimax_alpha_beta(board, depth, -math.inf, math.inf, True)
    return move,value
```

```python
def is_checkmate(chessboard, color):
    # Find the position of the King
    #print_board(chessboard)
    king_pos = None
    for i in range(len(chessboard)):
        for j in range(len(chessboard[i])):
            piece = chessboard[i][j]
            if piece != "   " and piece[1] == "K" and piece[2] == "K" and piece[0] == color:
                king_pos = (i, j)

    #print(king_pos,color)
    # Check if the King is in check
    from_row=king_pos[0]
    from_col=king_pos[1]
    to_row=king_pos[0]
    to_col=king_pos[1]


    # Check if the King is in check
    for i in range(len(chessboard)):
      for j in range(len(chessboard[i])):
        piece = chessboard[i][j]
        if piece != "   " and piece[0] != color[0]:
          #print("HERE",piece, i, j, king_pos[0], king_pos[1])
          if can_move(chessboard,piece, i, j, king_pos[0], king_pos[1]):
            #print("here",piece, i, j, king_pos[0], king_pos[1])
            return True

```

```python
def is_stalemate(chessboard, current_player):
   
    if not is_checkmate(chessboard, current_player):
        # all possible moves for the current player
        moves = get_all_moves(chessboard, current_player)
        # checking moves
        for move in moves:

            temp_chessboard = copy.deepcopy(chessboard)
            # Make the move on the copied chessboard
            move_piece(temp_chessboard, move[:3], move[3:])
            # Check if the current player is still not in check
            if not is_checkmate(chessboard, current_player):
                return False
        # if player has no moves tou it is a stalemate
        return True
    # checkmate case, it is not a stalemate
    return False
```
